```python
import pandas as pd
```

```python
fish=pd.read_csv('fish.csv')
fish
# class 1: 도미 / 0: 빙어
```

```python
bream=fish[fish['class']==1]
bream
```

```python
smelt=fish[fish['class']==0]
smelt
```

```python
bream_length=list(bream['length'])
bream_weight=list(bream['weight'])

smelt_length=list(smelt['length'])
smelt_weight=list(smelt['weight'])
# 병렬 계산이 되지 않게 일단 list로 변환 이따 array로 바꿀 예정
```

```python
print(bream_length)
print(smelt_length)
```

```python
print(bream_weight)
print(smelt_weight)
```

```python
# 문제집 : 길이, 무게 / 2차원 행렬
# 정답지 : 도미, 빙어 / 1차원 행렬
# 도미 35마리, 빙어 14마리
```

```python
length=bream_length + smelt_length
print(length)
```

```python
weight=bream_weight + smelt_weight
print(weight)
```

```python
data=[]

for i in range(49):
    data.append([length[i],weight[i]])
data
# 2차원 리스트 형태
```

```python
target=[1]*35+[0]*14
print(target)
```

```python
length=fish['length'].to_numpy()
weight=fish['weight'].to_numpy()
print(length)
print(weight)
```

```python
data=[]

for i in range(49):
    data.append([length[i],weight[i]])
data=np.array(data)
data
```

```python
target=fish['class'].to_numpy()
target
```

K 최근접 이웃 분류 모델

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier() # 모델 생성
knn.fit(data,target)
```

```python
knn.predict([[35,400]])
```

```python
knn.score(data,target)
# 실제값과 예측값이 완전히 일치하므로 1.0으로 나옴
# 예측값과 실체값의 맞은 비율
```

```python
target # 실제값
```

```python
knn.predict(data) # 예측값
```

```python
for i in range(1,50,2): # 가장 가까운 데이터를 골라서 다수결로 하므로 홀수개가 맞음
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(data,target)
    score=knn.score(data,target)
    print("이웃수 : {} // 점수 : {}".format(i,score))
# 29에서 갑자기 낮아지는 이유는 빙어데이터가 14개인데 나머지 15개를 도미데이터에서
# 가져오므로 29개부터는 빙어데이터를 다 도미데이터로 판별해서 틀린다고 봐야함
```
