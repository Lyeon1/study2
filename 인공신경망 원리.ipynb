```python
import matplotlib.pyplot as plt
import numpy as np

x1=np.array([2,4,6,8])
y=np.array([30,75,55,90])

a1=0
b=0
lr=0.005
# 처음엔 a랑 b값을 모르므로 아직 0

for i in range(1,3001):
    y_hat=a1*x1+b
    error=y_hat-y

    a1_diff=sum(2*error*x1) # a1의 미분값은 2*편차*x
    b_diff=sum(2*error*1)

    a1=a1-a1_diff * lr
    b=b-b_diff * lr

    print("{}회 학습 / 기울기 {} / 절편 {}".format(i,a1,b))
```

```python
plt.scatter(x1,y)
plt.plot(x1,a1*x1+b,'red') # y=ax+b 형태
```
