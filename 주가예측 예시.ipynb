```python
html=BeautifulSoup(url.text)

table=html.find('table')
pd.read_html(str(table))
```

```python
del table['전일비']
table
# 전일비가 가져오면서 +,- 가 사라져서 데이터가 오염되었다고 판단
```

```python
from tqdm import tqdm # for문의 진행도를 보는 라이브러리

total=[]

dic={'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 12_5_1)'}

for n in tqdm(range(1,662)):
    url=requests.get('https://finance.naver.com/item/sise_day.naver?code=005930&page={}'.format(n), headers=dic)
    html=BeautifulSoup(url.text)
    table=html.find('table')
    table=pd.read_html(str(table))[0].dropna()
    del table['전일비']
    total.append(table)
    # time.sleep(0.1)
```

```python
len(total)
```

```python
samsung=pd.concat(total, ignore_index=True)
samsung
```

```python
samsung.info()
# 날짜가 object로 되어있으니 문자 형태임
```

```python
samsung['날짜']=pd.to_datetime(samsung['날짜'])
samsung.info()
```

```python
samsung=samsung[::-1]
# 현재->과거 에서 과거->현재 순으로 바꿈
samsung
```

```python
samsung.index=range(len(samsung))
samsung
```

```python
data=[]
target=[]

for i in range(len(samsung)-1):
    a=list(samsung.iloc[i,1:])
    b=samsung.iloc[i+1,1]
    data.append(a)
    target.append(b)

import numpy as np

data=np.array(data)
target=np.array(target)
```

```python
from sklearn.ensemble import RandomForestRegressor # 회귀모델, 가격예측이니
import matplotlib.pyplot as plt

plt.plot(target)
```

```python
rf=RandomForestRegressor()
rf.fit(data,target)
```
